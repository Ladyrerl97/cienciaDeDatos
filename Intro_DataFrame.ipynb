# Introducción a Data Frames

Un data frame se puede definir como una variable que almacena tablas de datos, en lugar de números o palabras sencillas como las variables normales.Estas tablas se pueden manipular mediante pandas en python.

# Importación de bibliotecas

### Método import

Para importar bibliotecas o módulos en python, se hace uso del método "import"

```python
import pandas as pd
```

## Data Frame

### Método read_csv()

df_train=pd.read_csv("train.csv")

La línea de código anterior, indica que se esta creando una variable (Data Frame), llamada df_train, de pandas (pd), utilizando el método <b>read_csv</b> (método que lee datos de un archivo csv y los carga al DataFrame de pandas), entre parentesis y con comillas se escribe el nombre del archivo csv.

### Método .head

Este método permite visualizar determinado número de filas del data frame.

```python
df_train.head(20)
```

Para este caso heat(20) indica que se visualizarán 20 filas y 81 columnas.

### Atributo .shape

Este atributo permite conoces la <b>"forma" (shape)</b> de nuestro DataFrame.<br> (1460,81) indica 1460 filas por 81 columnas, es la cantidad de datos que contiene el DataFrame.

```python
df_train.shape
```

### Mostrar información de una columna en especifico

```python
df_train["Id"]
```

<b>nombre_DataFrame ["nombre de la columna a mostrar"]</b> <br> Esto permitirá mostar las primeras cinco y las últimas cinco filas de la tabla.

#### Mostrar varias columnas

Esto se logra iguakmente mediante el método anterior, pero al ser más de una columna, estas se deben de escribir como listas, es decir entre corchetes, por lo cual el método quedaría de la siguiente manera: <br> <b>nombre_DataFrame[["columna 1", "columna2]]</b>

```python
df_train[['Id', 'SalePrice']]
```

## Método mean

Este método permite obtener el promedio de una o varias columnas de Data Frame.

```python
df_train[["SalePrice", "Id"]].mean()
```

## Método describe

Es una herramienta que muestra todas las estadisticas de todo el DataFrame, como: <br> Recuento de valores no nulos<br> La media<br> La desviación estandar<br> Los valores máximos y mínimos<br> Varios porcentiles. 

```python
df_train.describe()
```
